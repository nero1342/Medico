```python
from datasets.image import ImageDataset
from torch.utils.data import DataLoader
```

```python
from torchmetrics.functional import accuracy, f1, iou
```

```python
from utils.config import Struct 
```

```python
import yaml
config = Struct(yaml.load(open('configs/pointrend.yaml', 'r'), Loader=yaml.Loader))
print(config)
```

```python
dataset_root = '../datasets/Test2020'
dataset = ImageDataset(dataset_root, subset='train')
```

```python
dataloader = DataLoader(dataset, batch_size = 1, shuffle = True, num_workers = 1, pin_memory=True)
```

```python
import matplotlib.pyplot as plt
# for data in dataloader:
data = dataset[0]
Fs = data['img'].cuda().unsqueeze(0)
Ms = data['mask'].cuda().unsqueeze(0)
plt.figure(figsize=(15, 15))
plt.subplot(131)
plt.imshow(Fs.permute(0, 2, 3, 1).detach().cpu()[0])
plt.subplot(132)
plt.imshow(Ms.detach().cpu()[0])
# break
```

```python
network = PointRend(config).cuda()
```

```python
pretrained_path = 'runs/Oct11_15.57.29_medico_PointRend/Oct11_15.57.29_medico_PointRend_4000.pth'
```

```python
import torch
network.load_state_dict(torch.load(pretrained_path))
```

```python

network.eval()

network.eval_()
```

```python
import matplotlib.pyplot as plt
import torch.nn.functional as F
torch.set_grad_enabled(False)
for i, data in enumerate(dataloader):
    print("Data ", i)
    Fs = data['img'].cuda()
    Ms = data['mask'].cuda()
    plt.figure(figsize=(15, 15))
    plt.subplot(131)
    plt.imshow(Fs.permute(0, 2, 3, 1).detach().cpu()[0])
    plt.subplot(132)
    plt.imshow(Ms.detach().cpu()[0])
    logits, history = network(Fs, Ms, 0)
    logits[:, 1] = logits[:, 1] * 5
    print(logits.shape, Ms.shape[-2:])
    logits = F.interpolate(logits, size = Ms.shape[-2:])
    
    preds = torch.argmax(F.softmax(logits, dim = 1), dim = 1)
    plt.subplot(133)        
    plt.imshow(preds.detach().cpu()[0])
    plt.show()
    
    acc = accuracy(preds, Ms)
    iou_score = iou(preds, Ms, reduction = 'none')[1]
    metrics = {
        "Accuracy": acc,
        "IoU": iou_score
    }
    print(metrics)
    
    for (id, sem_seg_logits, uncertainty_map, point_coords) in history:
        H, W = uncertainty_map.shape[-2:]
        point_coords = point_coords.reshape(-1, 2).detach().cpu()
        point_coords[:, 0] *= W
        point_coords[:, 1] *= H
        # print(point_coords.shape)
        cur_logits = F.interpolate(sem_seg_logits, size = logits.shape[2:])
        print(sem_seg_logits.shape, uncertainty_map.shape, cur_logits.shape)
        cur_preds = torch.argmax(F.softmax(cur_logits, dim = 1), dim = 1)
        acc = accuracy(cur_preds, Ms)
        iou_score = iou(cur_preds, Ms, reduction = 'none')[1]
        metrics = {
            "Accuracy": acc,
            "IoU": iou_score
        }
        print(metrics)
        plt.figure(figsize=(15, 15))
        plt.subplot(141)
        plt.imshow(Fs.permute(0, 2, 3, 1).detach().cpu()[0])
        plt.subplot(142)
        plt.imshow(cur_preds.detach().cpu()[0])
        plt.subplot(143)
        plt.imshow(uncertainty_map.detach().cpu()[0, 0])
        plt.subplot(144)
        plt.imshow(uncertainty_map.detach().cpu()[0, 0])
        plt.scatter(point_coords[:, 0], point_coords[:, 1], s = 1)
        plt.show()
    
    if i == 2:
        break
```
